When calculating a p-value, we imagine that the null hypothesis is true because we want to see how likely it is that wed get the results we did, just by chance, in that scenario. Essentially, were asking, If there really was no effect or no difference, how often would this kind of data show up? If the p-value is small, it means the results we got are pretty unlikely to happen by chance, suggesting that the null hypothesis probably isnt true. If the p-value is large, it means the results could easily happen even if the null hypothesis were true, so theres no strong reason to reject it.

Null Hypothesis (H0): The mean change is equal to 0.
Alternative Hypothesis (H0): The mean change is not equal to 0.
This test checks for extremes in both directions (positive and negative) and is suitable when the direction of the effect is unknown.
One-Tailed Test:
Null Hypothesis (H0): The mean change is equal to 0.
Alternative Hypothesis (H0): The mean change is greater than 0 (assuming we are testing for improvement).
This test focuses only on the upper tail, checking if the observed statistic is significantly greater than what would be expected under the null hypothesis.

We modified the existing code to compute a p-value for a one-tailed hypothesis test. The change involved adjusting the condition to check if the simulated statistics are greater than or equal to the observed statistic, rather than checking both tails.

#### Introduction 
In this experiment, we set out to determine whether students in STA130 can correctly identify whether milk or tea was poured first in a cup of tea. This study draws inspiration from the famous experiment conducted by Ronald Fisher and Dr. Muriel Bristol in the 1920s. Bristol claimed she could discern the order in which milk and tea were poured, sparking Fishers skepticism and leading him to conduct a test. Our goal is to assess the performance of 80 STA130 students in a similar taste test to see if they share Bristols perceptive ability.

- While Fishers original experiment involved a smaller sample and a more personal context, our study examines a larger group of STA130 students, making it a bit more abstract. This difference highlights the potential variability in taste perception and whether individuals can reliably discern the pouring order based solely on their experiences or preferences.

#### Null Hypothesis (H0)
We start by assuming that students are merely guessing when they determine the order of pouring, meaning their success rate would be no better than flipping a coin.

#### Alternative Hypothesis (HA)
The proportion of STA130 students who can correctly identify the order of pouring is greater than the chance level, which is p > 0.5

```python
import scipy.stats as stats

# Given values
p_0 = 0.5  # null hypothesis proportion
n = 80     # sample size
x = 49     # number of correct identifications
p_hat = x / n  # observed proportion

# Calculate the standard error
SE = (p_0 * (1 - p_0) / n) ** 0.5

# Calculate the z-score
Z = (p_hat - p_0) / SE

# Calculate the p-value for a one-tailed test
p_value = 1 - stats.norm.cdf(Z)

Z, p_value
```

```python
import matplotlib.pyplot as plt

# Data
correct = 49
incorrect = 80 - correct
labels = ['Correct', 'Incorrect']
values = [correct, incorrect]

# Create bar chart
plt.figure(figsize=(8, 5))
plt.bar(labels, values, color=['green', 'red'])
plt.title('Correct vs. Incorrect Responses in Tea Experiment')
plt.ylabel('Number of Students')
plt.ylim(0, 80)

# Adding value labels on top of bars
for i, value in enumerate(values):
    plt.text(i, value + 1, str(value), ha='center')

plt.show()
```

```python
import numpy as np
import seaborn as sns

# Generate x values for the normal distribution
x = np.linspace(0, 1, 100)
y = stats.norm.pdf(x, p_0, SE)

# Create the plot
plt.figure(figsize=(10, 6))
sns.lineplot(x=x, y=y, color='blue', label='Normal Distribution')
plt.axvline(x=p_hat, color='orange', linestyle='--', label='Observed Proportion (p_hat)')
plt.axvline(x=p_0, color='red', linestyle='--', label='Null Hypothesis Proportion (p_0)')

plt.title('Normal Distribution of Proportions with Z-Score')
plt.xlabel('Proportion of Correct Responses')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# Plot the normal distribution and indicate z-score
plt.figure(figsize=(10, 6))
sns.kdeplot(data=np.random.normal(loc=p_0, scale=SE, size=1000), fill=True, color='blue', label='Normal Distribution')
plt.axvline(x=p_hat, color='orange', linestyle='--', label='Observed Proportion (p_hat)')
plt.axvline(x=p_0, color='red', linestyle='--', label='Null Hypothesis Proportion (p_0)')
plt.axvline(x=p_0 + Z * SE, color='purple', linestyle='--', label='Z-Score')

plt.title('Normal Distribution with Observed Proportion and Z-Score')
plt.xlabel('Proportion of Correct Responses')
plt.ylabel('Density')
plt.legend()
plt.show()
```

If our p-value is less than or equal to 0.05, we would conclude that there is significant evidence to reject the null hypothesis, suggesting that STA130 students can indeed distinguish between the pouring order better than chance. On the other hand, if the p-value is greater than 0.05, we fail to reject the null hypothesis, indicating that their performance does not significantly exceed random guessing.
